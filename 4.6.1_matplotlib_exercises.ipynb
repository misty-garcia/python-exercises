```python
import matplotlib.pyplot as plt
import math
import random
%matplotlib inline
```

```python
x = range(-50,50)
y = [n**2 - n + 2 for n in x]
 
plt.text(0,100, '(0,0)', color = 'red')
plt.plot(x,y)
plt.show()
```

```python
x = range(2,100)
y = [n**(1/2) for n in x]

plt.plot(x,y)
plt.title("The square root of $x$")
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.show()
```

```python
x = range(-100,100)
y = [n**3 for n in x]

plt.plot(y)
plt.title("$x$ cubed")
plt.xlabel('$x$')
plt.ylabel('$x$^3')
plt.show()
```

```python
x = range(-100,100)
y = [math.tan(n) for n in x]

plt.plot(y)
plt.title("The tangent of $x$")
plt.xlabel('$x$')
plt.ylabel('tan($x$)')
plt.show()
```

```python
x = range(0,10)
y = [2**n for n in x]

plt.plot(x,y)
plt.title("2 to the $x$")
plt.xlabel('$x$')
plt.ylabel('2^($x$)')
plt.show()
```
