### Find 5 letter words whose alphabetic number of the last four characters total to the value of the first. 

```python
with open('/usr/share/dict/words') as f:
    words = f.read().split('\n')
print(words[:5])
```

```python
def alphabet_number(letter):
    return ord(letter.upper()) - 64
```

```python
alphabet_number("a")
```

```python
list_words = []
for x in words:
    if len(x) == 5:
        list_words.append(x)
list_words
```

```python
for x in list_words:
    total = 0
    count = 0
    for y in x:
        if count == 0:
            total = alphabet_number(y)
        else:
            total -= alphabet_number(y)
        count += 1
    if total == 0:
        print(x)
```

```python
print(alphabet_number("r"))
print(alphabet_number("a"))
print(alphabet_number("f"))
print(alphabet_number("f"))
print(alphabet_number("e"))
```
